# Monte Carlo method

## 0. Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

## 1. Introduction

In the following notebook we will see how we can estimate $\pi$ with a random sampling method. <br>
Suppose we have a square of $2L$ unit length, within this square a circle is drawn, **radius** of the circle is indeed of $L$ unit lentgh. Since estimating the area of a square is easy, $ 2L\times 2L$, if we throw several rocks (with neglectable size compare to $L$) the fraction of rocks which land into the circle can be interperet as the area of the circle. <br>
$$
\frac{S_{circle}}{S_{Square}} = \frac{\pi \times (L)^2}{(2L)^2} \;\;\;\; (1.1).
$$
If we simplify the above expression we will have:
$$
\frac{S_{circle}}{S_{Square}} = \frac{\pi}{4}.
$$
<br> Hence if we throw a large number of rocks, we can estimate $\pi$ as follows:
$$
\pi = 4 \times \frac{No.\;rocks\;landed\;into\;the\;circle}{No.\;rocks\;landed\;in\;the\;sqaure} \;\;\;\; (1.2).
$$
<br>Indeed "No. rocks landed in the sqaure" is **the total number of rocks**.

## 2. Evaluating $\pi$ program

In the following, we implement an algorithm to estimate $\pi$: <br>
1. Set up boundary by defining $L$;<br>
2. Generate random points individually for x and y in $[-L,L]$; <br>
3. Calculate the distance of random point from the origin ($ r^2 = x^2 + y^2$); <br>
4. If $ r^2 \leq (L)^2 $ count the random point as within the circle; <br>
5. Repeat 2-4 steps for sufficient number, i.e., "runs"; <br>
6. Calculate $\pi$ with (1.2) expression. 

```python
# 1. defining L and number of runs
L = 2
runs  = 1000
Circle_hit = 0
Total_hit = 0

rc_squared = L**2 # To avoid repeating identical arithemic operations

for _ in range(runs):
    # 2. Selecting random x,y
    x = np.random.uniform(-L,L)
    y = np.random.uniform(-L,L)
    # 3. calculating r^2 of the random point
    r_squared = x**2 + y**2 
    # 4. Check whether it is a successful hit or not
    if ( r_squared <= rc_squared):
        Circle_hit += 1
    Total_hit +=1
    # 5. looping for "runs" number

# 6. Calculating Pi
Pi = 4*(Circle_hit/Total_hit)

print("Estimated pi is: ", Pi)
```

**Lets illustrate this simulation:**

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
L = 2
runs  = 10000
Circle_hit = 0
Total_hit = 0

rc_squared = L**2 

# drawing the circle
theta = np.linspace( 0 , 2 * np.pi , 150 ) 
radius = L
a = radius * np.cos( theta )
b = radius * np.sin( theta )
ax.plot(a,b, color='black' , lw=5)
ax.set_aspect('equal')
ax.set_xlim(-L,L)
ax.set_ylim(-L,L)


for _ in range(runs):
    x = np.random.uniform(-L,L)
    y = np.random.uniform(-L,L)
    r_squared = x**2 + y**2 
    if ( r_squared <= rc_squared):
        Circle_hit += 1
        ax.scatter(x,y, color='green', s= 20, edgecolor='black')
    else:
        ax.scatter(x,y, color='red', s= 20, edgecolor='black')  
    Total_hit +=1

    
plt.show()
Pi = 4*(Circle_hit/Total_hit)

print(Pi)
```
